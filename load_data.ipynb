```python
import overpass
api = overpass.API()
queries = ['"shop"="supermarket"', '"amenity"="hospital"']

dfs = []
for query in queries:
    response = api.Get('area["name"="Уфа"];(node[{}](area););out;'.format(query))
    dfs.append(gpd.GeoDataFrame.from_features(response['features']))
```

```python
pd.concat(dfs, axis=0)
```
